```python
# Warmup: How many blocks of code does this have?

y = 7
z = True

# This is an example of "chained" conditional that is "nested" inside another conditional.
if z:
    print('z is True.')
    if y < 7:
        print('y is less than 7.')
    elif y == 7:                      # elif -> chained
        print('y is equal to 7.')
    else:
        print('y is greater than 7')
else:
    print('z is False.')
    print('Not looking at y.')
print('Done!')
    
```

## Finding the Maximum Value

```python
# Example 3a: define the function
def f(x):
    return 5 - (x-2)**2
```

```python
# ... and test it!
print(f(-1))
print(f(0))
print(f(3))
print(f(6))
```

```python
# Example 3b_QUESTION: Write code to iterate over values 
# We'll write this one together
```

```python
# Example 3b_ANSWER: Write code to iterate over values 

start_x = 0
end_x = 5
delta_x = 0.1

max_y = f(start_x)
max_x = start_x

current_x = start_x

while current_x <= end_x:
    current_y = f(current_x)
    if current_y > max_y:
        max_x = current_x
        max_y = current_y
    current_x += delta_x
print("max occurred at", max_x, max_y)
```

```python
# Example 4: Riemann Sum
# Copy/paste the code above
# Change the block inside the while loop to find area under the curve
# Do we have to change the condition?
start_x = 1
end_x = 5
delta_x = 0.2

total_area = 0
current_x = start_x

print("total area is: ", total_area)
```

```python
# Example 4: Riemann Sum SOLUTION
# Copy/paste the code above
# Change the block inside the while loop to find area under the curve
# Do we have to change the condition?
start_x = 1
end_x = 5
delta_x = 0.2

total_area = 0
current_x = start_x

# How can we count the number of iterations to make sure we aren't off by 1?
while current_x < end_x:
    current_y = f(current_x)
    total_area += current_y * delta_x # height * width of rectangle
    current_x += delta_x
    
print("total area is: ", total_area)
```

```python
# Example 5:  write a function to determine if an integer is prime
import math

def is_prime(x):
    """ returns True if x is prime, false otherwise. Assumes x is an int"""
    if x <= 1:
        return False
    
    divisor = 2

    while divisor <= (x - 1): # for a quicker solution, we could check math.sqrt(x). Note! We have to be careful of the case where x = 2. For details, see https://stackoverflow.com/questions/5811151/why-do-we-check-up-to-the-square-root-of-a-prime-number-to-determine-if-it-is-pr
        if x % divisor == 0:  # its divisible
            return False      # not prime
        divisor += 1          # try the next number up
    return True               # if we made it to this point, it isn't divisble by any number!

print(is_prime(101))
print(is_prime(36))
print(is_prime(18))
print(is_prime(7))
print(is_prime(2))
print(is_prime(1))
```
