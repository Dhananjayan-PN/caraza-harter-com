```python
# The / operator does Division as a float. Try a few examples
print(55/5)
print(2/19)
print(7/2)
```

```python
# The // operator does Integer division. Try a few examples
print(55//5)
print(2//19)
print(7//2)
```

```python
# Sometimes / will display small numbers in scientific notation
# print out calcualtions that result in very small numbers
print(34 / 1000000)
```

```python
# the modulus operator (%) gives the Integer Remainder
print( 15 % 7)
# try other examples
print( 15 % 3)
```

```python
# use the modulus along with Integer Division to grab a certain digit of an integer
print((6834//100) % 100% 10)
```

Things to know about Strings


```python
# Strings can be enclosed in '' or ""
print('hello')
print("hello")
```

```python
# Strings can contain special escape characters that start with \
# \n is for newline
# \t is for tab
# \" is for a literal quotation mark
print("he\nll\to")
```

```python
# the + operator concatenates two strings
"hello" + "There"
```

```python
# the * operator repeats a string
"ha " * 10
```

Boolean Operators




```python
# try out and, or, not with True and False values
print(True and False)
print(False or True)
```

```python
# try comparison operators
print(3 < 7)
print(4 <= 4)
```

```python
# try to compare two strings with < or >  or >= or <=
"a" > "A"
```

```python
# practice using != (not equal) and == (is equal)
print(45 != 40 + 1)
print(45 == 40 + 1 )
```

```python
# compare two types for equality
print(type(4.5))
print(type("hello"))
print(type(2))
print(type(True))
```

```python
# compound comparisons with and
print (4<7 and 19 > 2)
```

```python
# compound comparisons with or
print (4<7 or 19 > 2)
```
