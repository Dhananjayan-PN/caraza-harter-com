```python
# Example 4a: define the function and test it
def f(x):
    return 5 - (x-2)**2
f(3)
```

```python
# Example 4b: write code to iterate over values 

start_x = 1
end_x = 5
delta_x = 0.1

max_y = f(start_x)
max_x = start_x
print(max_y)

current_x = start_x

while current_x <= end_x:  # condition
    y = f(current_x)
    print(current_x, y)
    if y > max_y:       # found the max y
        max_y = y       # found the max x
        max_x = current_x
    current_x += delta_x
print("max occurred at", max_x, max_y)
```

```python
# Example 4c: copy/paste the code above
# change the block intide the while loop to find area under the curve
start_x = 1
end_x = 5
delta_x = 0.2

total_area = 0
current_x = start_x

while current_x <= end_x:  # condition
    y = f(current_x)
    print(current_x, y)
    
    area = y * delta_x  # area of one rectangle
    total_area += area
    current_x += delta_x

    
print("total area is: ", total_area)
```

```python
# Example 5:  write a function to determine of an integer is prime
import math

def is_prime(x):
    """ returns True if x is prime, false otherwise. Assumes x is positive"""
    # check for numbers that evenly go into x
  
    # start at 2
    divisor = 2
    # while divisor is <= square root of number
    
    while divisor <= math.sqrt(x):
        if x % divisor == 0: # its divisble
            return False     # not prime
        divisor += 1         # try the next number up
    return True

is_prime(101)
```

```python
is_prime(36)
```

```python
is_prime(2) #2 is prime, this should return True
```

```python
# Example 6:  Write a loop that calls is_prime for all integers from 2 to 100

print("Prime numbers:")
number = 2
while number <= 100:
    if is_prime(number):
        print(number, "is prime")
    else:
        print(number, "is not prime")
    number += 1
```
