```python
# Warmup 3:  Fix the errors in this version of is_prime

import math

def is_prime(x):
    """ returns True if x is prime, false otherwise. Assumes x is positive"""
    
    divisor = 2
    while divisor <= math.sqrt(x):
        if x % divisor == 0: # its divisble
            return True     # not prime
        divisor += 1         # try the next number up
    return False

print(is_prime(12)) # False
print(is_prime(13)) # True
print(is_prime(55)) # False
```

#### Acey-Deucey Game: An application of break, continue, and nested loops

https://en.wikipedia.org/wiki/Acey_Deucey_(card_game)



```python
# Acey Deucey Game

import random

money = 10
num_rounds = 5
current_round = 1

print("welcome to acey deucey")
print("you have: $", money)

while current_round <= num_rounds:
    card1 = random.randint(1,13)
    card2 = random.randint(1,13)
    # if the cards are equal, skip all the code below and go back to the top
    
    print("your cards: ", card1,"\t", card2)
    if card1 == card2:
        continue
    
    wager = int(input("how much do you want to wager? "))  
    while wager > money or wager < 0:
        print("you cannot bet that amount")
        wager = int(input("how much do you want to wager? ")) 
        
    
    card3 = random.randint(1,13)
    print("the middle card was" , card3)
    if wager != 0:
        if is_between(card1, card3, card2) :
            print('you win!')
            money += wager
        else:
            print('you lose!')
            money -= wager

    # if you are out of money, leave the loop
    print("you have: $", money)
    if money <= 0:
        break
    current_round += 1

print("goodbye")
```
