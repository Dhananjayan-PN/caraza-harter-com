```python
# now let's try to dog function....you do this on your own

# import this module
import dog
# help
help(dog)
# try each function
print(dog.fetch())
print(dog.speak())
```

```python
# potential bug:  if we use from cat import speak, from dog import speak
from cat import speak
from dog import speak
print(speak())
print(cat.speak())
print(dog.speak())
```

```python
# how can we fix this? 
import cat
import dog
print(dog.fetch())
```

```python
# key takeway....using import dog, and then calling dog.speak() is safest
```

```python
def add3(x,y=100,z=100):  # x, y, and z are parameters
    """adds three numbers""" #documentation string
    print ("y=" + str(y))
    return x+y+z

add3(100,10,5) #positional arguments
```

```python
print(add3(4,2,1)**2)
```

```python
add3(x=1, z=2, y=5) #keyword arguments
```

```python
add3(5,6) #using a default argument....must be at the end of the parameter list
# z=100 is the deafult argument
```

```python
# incorrect use of arguments in function calls
add3(5, x=2)
```

```python
# other incorrect uses
add3(z=3, 5, 7)
```
